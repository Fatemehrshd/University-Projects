## Low-pass filtering Example

```python
from scipy.signal import butter, lfilter
import numpy as np
import matplotlib.pyplot as plt

def butter_lowpass(data, cutoff_freq, sampling_rate, order=4):
    nyquist = 0.5 * sampling_rate
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    filtered_data = lfilter(b, a, data)
    return filtered_data

# Example usage:
sampling_rate = 1000  # Hz
cutoff_frequency = 50  # Hz
time = np.arange(0, 1, 1/sampling_rate)
signal = np.sin(2 * np.pi * 100 * time) + 0.5 * np.random.normal(size=len(time))
filtered_signal = butter_lowpass(signal, cutoff_frequency, sampling_rate)

# Plot the original and filtered signals
plt.plot(time, signal, label='Original Signal')
plt.plot(time, filtered_signal, label='Filtered Signal')
plt.legend()
plt.show()
```

## Burned Calories formula

```python
def calculate_calories_per_minute_men(HR, W, A):
    return (-55.0969 + (0.6309 * HR) + (0.1988 * W) + (0.2017 * A)) / 4.184

def calculate_calories_per_minute_women(HR, W, A):
    return (-20.4022 + (0.4472 * HR) - (0.1263 * W) + (0.074 * A)) / 4.184
```

```python
def calculate_calories_per_minute_with_VO2max_men(HR, VO2max, W, A):
    return (-95.7735 + (0.634 * HR) + (0.404 * VO2max) + (0.394 * W) + (0.271 * A)) / 4.184

def calculate_calories_per_minute_with_VO2max_women(HR, VO2max, W, A):
    return (-59.3954 + (0.45 * HR) + (0.380 * VO2max) + (0.103 * W) + (0.274 * A)) / 4.184
```
